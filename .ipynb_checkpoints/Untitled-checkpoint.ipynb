```python
import pandas as pd

# load historical price data
data = pd.read_csv('btcusdt.csv', parse_dates=['timestamp'], index_col='timestamp')

# define the weights for the three strategies
mean_reversion_weight = 0.4
trend_following_weight = 0.3
breakout_weight = 0.3

# define the parameters for the mean reversion strategy
mean_reversion_periods = 10

# define the range for the breakout strategy
range_high = 50000
range_low = 40000

# initialize variables for tracking trades and P&L
position = 0
entry_price = 0
trades = []
pnl = 0

# loop through each row of the historical price data
for i, row in data.iterrows():

    # calculate the signals for each strategy
    mean_reversion_signal = 0
if (i - mean_reversion_periods * pd.Timedelta(days=1)) >= pd.Timedelta(0):
        mean = data['close'].iloc[i-mean_reversion_periods:i].mean()
        std = data['close'].iloc[i-mean_reversion_periods:i].std()
        if row['close'] < (mean - (2 * std)):
            mean_reversion_signal = 1
        if row['close'] > (mean + (2 * std)):
            mean_reversion_signal = -1

        trend_following_signal = 0
        if i > 0 and row['close'] > data['close'].iloc[i-1]:
            trend_following_signal = 1
        if i > 0 and row['close'] < data['close'].iloc[i-1]:
            trend_following_signal = -1

        breakout_signal = 0
        if row['close'] > range_high:
            breakout_signal = 1
        if row['close'] < range_low:
            breakout_signal = -1

    # calculate the weighted average signal
        weighted_signal = (mean_reversion_weight * mean_reversion_signal) + (trend_following_weight * trend_following_signal) + (breakout_weight * breakout_signal)

    # close existing position if signal changes from buy to sell or vice versa
        if weighted_signal > 0 and position == -1:
            trades.append({'timestamp': i, 'type': 'Sell', 'price': row['close'], 'size': abs(position)})
            pnl += abs(entry_price - row['close'])
            position = 0
        elif weighted_signal < 0 and position == 1:
            trades.append({'timestamp': i, 'type': 'Buy', 'price': row['close'], 'size': abs(position)})
            pnl += abs(entry_price - row['close'])
            position = 0

        # open new position if signal is buy or sell
        if weighted_signal > 0 and position == 0:
            entry_price = row['close']
            trades.append({'timestamp': i, 'type': 'Buy', 'price': entry_price, 'size': 0.01})
            position = 1
        elif weighted_signal < 0 and position == 0:
            entry_price = row['close']
            trades.append({'timestamp': i, 'type': 'Sell', 'price': entry_price, 'size': 0.01})
            position = -1

# print the list of trades and the total P&L
print(trades)
print('Total P&L:', pnl)
```

```python
import pandas as pd

# Load the data
df = pd.read_csv('btcusdt.csv')

# Define the initial capital
initial_capital = 10000

# Define the indicators
df['sma_50'] = df['close'].rolling(window=50).mean()
df['sma_200'] = df['close'].rolling(window=200).mean()

# Define the trading strategy function
def trading_strategy(df):
    buy_signals = []
    sell_signals = []
    position = 0

    for i in range(len(df)):
        if df['sma_50'][i] > df['sma_200'][i] and position == 0:
            buy_signals.append(df['close'][i])
            sell_signals.append(float('nan'))
            position = 1
        elif df['sma_50'][i] < df['sma_200'][i] and position == 1:
            buy_signals.append(float('nan'))
            sell_signals.append(df['close'][i])
            position = 0
        else:
            buy_signals.append(float('nan'))
            sell_signals.append(float('nan'))
    
    if position == 1:
        buy_signals.append(float('nan'))
        sell_signals.append(df['close'][len(df)-1])
    else:
        buy_signals.append(float('nan'))
        sell_signals.append(float('nan'))

    return buy_signals, sell_signals

# Get the buy and sell signals
buy_signals, sell_signals = trading_strategy(df)

# Create a new dataframe with the buy and sell signals
signals = pd.DataFrame(index=df.index)
signals['buy_signals'] = buy_signals
signals['sell_signals'] = sell_signals

# Define the backtesting function
def backtest(df, signals, initial_capital):
    positions = pd.DataFrame(index=signals.index).fillna(0.0)
    positions['position'] = signals['buy_signals'].fillna(method='ffill')
    positions['position'] = positions['position'].fillna(0.0)
    positions['position'] = positions['position'].apply(lambda x: 1 if x > 0 else -1)

    portfolio = positions.multiply(df['close'], axis=0)
    pos_diff = positions.diff()

    portfolio['holdings'] = (positions.multiply(df['close'], axis=0)).sum(axis=1)
    portfolio['cash'] = initial_capital - (pos_diff.multiply(df['close'], axis=0)).sum(axis=1).cumsum()

    portfolio['total'] = portfolio['cash'] + portfolio['holdings']
    portfolio['returns'] = portfolio['total'].pct_change()

    return portfolio

# Run the backtesting function and get the portfolio
portfolio = backtest(df, signals, initial_capital)

# Print the final portfolio value
print(f"Final portfolio value: {portfolio['total'][-1]}")
```

```python
import pandas as pd
import numpy as np

# Load data from CSV file
data = pd.read_csv('btcusdt.csv', index_col='timestamp', parse_dates=True)

# Calculate 50-day and 200-day simple moving averages
data['SMA_50'] = data['close'].rolling(window=50).mean()
data['SMA_200'] = data['close'].rolling(window=200).mean()

# Generate trading signals based on moving average crossover
data['signal'] = np.where(data['SMA_50'] > data['SMA_200'], 1, -1)

# Calculate hourly returns based on trading signals
data['return'] = data['close'].pct_change(1) * data['signal'].shift(1)

# Calculate cumulative returns
data['cumulative_return'] = (1 + data['return']).cumprod()

# Print final cumulative return
print('Cumulative Return:', data['cumulative_return'][-1])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('btcusdt.csv')

# Set the index to the datetime column
df['datetime'] = pd.to_datetime(df['datetime'])
df.set_index('datetime', inplace=True)

# Define the mean reversion strategy
def mean_reversion_strategy(df, lookback_periods, entry_threshold, exit_threshold):
    # Calculate rolling mean and standard deviation
    rolling_mean = df['close'].rolling(window=lookback_periods).mean()
    rolling_std = df['close'].rolling(window=lookback_periods).std()

    # Calculate upper and lower bands
    upper_band = rolling_mean + (rolling_std * entry_threshold)
    lower_band = rolling_mean - (rolling_std * entry_threshold)

    # Initialize the position
    position = 0

    # Create a list to store the P&L
    pnl_list = []

    # Loop over the data
    for i in range(len(df)):
        # Check if the price is above the upper band
        if df['close'][i] > upper_band[i]:
            if position == 0:
                position = -1
                entry_price = df['close'][i]
        # Check if the price is below the lower band
        elif df['close'][i] < lower_band[i]:
            if position == 0:
                position = 1
                entry_price = df['close'][i]
        # Check if the position should be closed
        elif abs(df['close'][i] - entry_price) >= exit_threshold:
            if position == -1:
                pnl = entry_price - df['close'][i]
                pnl_list.append(pnl)
                position = 0
            elif position == 1:
                pnl = df['close'][i] - entry_price
                pnl_list.append(pnl)
                position = 0

    # Calculate the total P&L
    total_pnl = np.sum(pnl_list)

    # Calculate the average P&L per trade
    if len(pnl_list) > 0:
        avg_pnl_per_trade = total_pnl / len(pnl_list)
    else:
        avg_pnl_per_trade = 0

    # Return the total P&L and the average P&L per trade
    return total_pnl, avg_pnl_per_trade

# Set the strategy parameters
lookback_periods = 50
entry_threshold = 2
exit_threshold = 1

# Run the strategy and get the results
total_pnl, avg_pnl_per_trade = mean_reversion_strategy(df, lookback_periods, entry_threshold, exit_threshold)

# Print the results
print('Total P&L: {:.2f}'.format(total_pnl))
print('Avg P&L per trade: {:.2f}'.format(avg_pnl_per_trade))

# Plot the data
plt.plot(df.index, df['close'])
plt.plot(df.index, df['close'].rolling(window=lookback_periods).mean(), label='Rolling Mean')
plt.plot(df.index, df['close'].rolling(window=lookback_periods).mean() + (df['close'].rolling(window=lookback_periods).std() * entry_threshold), label='Upper Band')
plt.plot(df.index, df['close'].rolling(window=lookback_periods).mean() - (df['close'].rolling(window=lookback_periods).std() * entry_threshold), label='Lower Band')
plt.legend()
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data from CSV file
df = pd.read_csv('btcusdt.csv', parse_dates=['timestamp'], index_col='timestamp')

# Set parameters for strategy
lookback_period = 24 # number of hours to look back for calculating rolling mean and standard deviation
entry_z_score = -2.0 # entry threshold for z-score
exit_z_score = 0.0 # exit threshold for z-score
stop_loss = 0.05 # stop loss percentage

# Calculate rolling mean and standard deviation
df['rolling_mean'] = df['close'].rolling(window=lookback_period).mean()
df['rolling_std'] = df['close'].rolling(window=lookback_period).std()

# Calculate z-score
df['z_score'] = (df['close'] - df['rolling_mean']) / df['rolling_std']

# Set up variables for trade tracking
position = 0
entry_price = 0
stop_loss_price = 0
pnl = []

# Loop through each row and generate signals and track pnl
for i, row in enumerate(df.itertuples()):
    # Check if there's an open position and if stop loss has been hit
    if position != 0 and row.close <= stop_loss_price:
        pnl.append((stop_loss_price - entry_price) * position)
        position = 0
        entry_price = 0
        stop_loss_price = 0
    # Check for entry signal
    elif row.z_score <= entry_z_score and position == 0:
        position = 1
        entry_price = row.close
        stop_loss_price = entry_price * (1 - stop_loss)
    # Check for exit signal
    elif row.z_score >= exit_z_score and position == 1:
        pnl.append((row.close - entry_price) * position)
        position = 0
        entry_price = 0
        stop_loss_price = 0

# Calculate total P&L and print summary statistics
total_pnl = sum(pnl)
winning_trades = sum([pnl_i > 0 for pnl_i in pnl])
losing_trades = sum([pnl_i < 0 for pnl_i in pnl])
total_trades = len(pnl)
win_rate = winning_trades / total_trades
average_pnl = total_pnl / total_trades
print('Total P&L: {:.2f}'.format(total_pnl))
print('Win Rate: {:.2%}'.format(win_rate))
print('Average P&L: {:.2f}'.format(average_pnl))

# Plot P&L over time
plt.plot(np.cumsum(pnl))
plt.xlabel('Trade')
plt.ylabel('Cumulative P&L')
plt.title('Mean Reversion Strategy P&L')
plt.show()
```

```python
import pandas as pd
import numpy as np

# Load data from CSV file
data = pd.read_csv('btcusdt_15m.csv', index_col='timestamp', parse_dates=True)

# Calculate 50-day and 200-day simple moving averages
data['SMA_50'] = data['close'].rolling(window=50).mean()
data['SMA_200'] = data['close'].rolling(window=200).mean()

# Generate trading signals based on moving average crossover
data['signal'] = np.where(data['SMA_50'] > data['SMA_200'], 1, -1)

# Calculate hourly returns based on trading signals
data['return'] = data['close'].pct_change(1) * data['signal'].shift(1)

# Calculate cumulative returns
data['cumulative_return'] = (1 + data['return']).cumprod()

# Print final cumulative return
print('Cumulative Return:', data['cumulative_return'][-1])
```

```python
import pandas as pd
import numpy as np
from datetime import datetime
from time import time

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('btcusdt_15m.csv')

# Set the timestamp column as the index
df.set_index('timestamp', inplace=True)

# Define the trading parameters
window_size = 20
buy_threshold = 2.0
sell_threshold = -2.0
quantity = 0.01  # Replace with your own quantity

# Define the initial state of the trading strategy
position = None
buy_price = None
sell_price = None
profit = 0.0
trades = []

# Define a function to calculate the moving average and standard deviation of a price series
def calculate_stats(prices):
    moving_average = np.convolve(prices, np.ones(window_size)/window_size, mode='valid')
    standard_deviation = np.std(prices[-window_size:])
    return moving_average, standard_deviation

# Define a function to place a buy order
def buy(timestamp, price):
    global position, buy_price
    order = {'timestamp': timestamp, 'side': 'buy', 'price': price, 'quantity': quantity}
    trades.append(order)
    position = 'long'
    buy_price = price

# Define a function to place a sell order
def sell(timestamp, price):
    global position, sell_price, profit
    order = {'timestamp': timestamp, 'side': 'sell', 'price': price, 'quantity': quantity}
    trades.append(order)
    position = None
    sell_price = price
    profit += (sell_price - buy_price) * quantity

# Loop through each row of the DataFrame and apply the trading strategy
start_time = time()
for index, row in df.iterrows():
    price = row['close']
    moving_average, standard_deviation = calculate_stats(df.loc[:index, 'close'].values)

    if position == 'long':
        if price > moving_average + sell_threshold * standard_deviation:
            sell(index, price)
    else:
        if price < moving_average + buy_threshold * standard_deviation:
            buy(index, price)

# Print the results of the backtest
end_time = time()
print('Backtest results:')
print('-----------------')
print(f'Total profit: {profit:.2f} USDT')
print(f'Number of trades: {len(trades)}')
print(f'Average time per trade: {(end_time - start_time) / len(trades):.2f} seconds')
```
